```python
# Шаг 1: Импортируйте данные
data = pd.read_csv('GAZP.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Предполагаем, что у вас есть столбец 'Close' с ценами закрытия
data['Returns'] = data['Close'].pct_change()

# Удаляем NaN значения
returns = data['Returns'].dropna()

# Шаг 2: Определение параметров
confidence_level = 0.95
num_simulations = 10000

# Шаг 3: Реализация метода Монте-Карло
mean_return = returns.mean()
std_dev = returns.std()

# Генерация случайных доходностей
simulated_returns = np.random.normal(mean_return, std_dev, (num_simulations, len(returns)))

# Рассчитываем конечные стоимости портфеля (предполагаем начальную стоимость $1)
initial_investment = 1
final_values = initial_investment * (1 + simulated_returns).cumprod(axis=1)

# Определяем LVaR
losses = initial_investment - final_values[:, -1]
LVaR = np.percentile(losses, (1 - confidence_level) * 100)

print(f"LVaR на уровне доверия {confidence_level*100}% составляет: {LVaR:.2f}")
```
