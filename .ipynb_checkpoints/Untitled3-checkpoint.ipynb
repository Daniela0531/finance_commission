```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# VaR на уровне 95% за день показывает, 
# какую долю стоимости актив может потерять 
# за день с вероятностью 5% 
# (то есть в 1 из 20 дней убыток превысит это значение)
alpha=0.05
```

```python
# Предобработка данных
```

```python
files_data=dict.fromkeys(['GAZP.csv', 'SBER.csv', 'MGNT.csv'])
```

```python
for file_name in files_data:
    files_data[file_name]=pd.read_csv(file_name)
```

```python
for company in files_data.values():
  company['Изм. %'] = [float(val[:-1].replace(',', '.'))/100 for val in company['Изм. %']]
```

```python
period = len(files_data['GAZP.csv'])
sorted_data = dict.fromkeys(files_data.keys())

for company, data in files_data.items():
    sorted_data[company] = np.sort(data['Изм. %'])
```

```python
# Cтандартный VaR
```

```python
coef = int(period * alpha)

for company, profit in sorted_data.items():
    var_ = float(profit[coef] * 100)
    print('VaRst_{name} = {var}%'.format(name=company, var=-var_))
```

```python
# График доходности
```

```python
for file_name in files_data.keys():
    cut_cases = sorted_data[file_name][:coef]
    remain_cases = sorted_data[file_name][coef:len(sorted_data[file_name])]
    
    plt.figure(figsize=(15, 10))
    plt.title(file_name.rsplit(".csv"))
    plt.plot(range(len(cut_cases)), cut_cases, color='red')
    plt.plot(range(len(cut_cases), len(remain_cases) + len(cut_cases)), remain_cases, color='green')
    plt.xlabel('Дни, шт')
    plt.ylabel('Доходность, %')
    plt.grid()
    plt.show()
```

```python
# Риск портфеля
```

```python
for company in files_data.values():
  company['Цена'] = [float((price.replace('.', '')).replace(',', '.')) for price in company['Цена']]
```

```python
portfolio_price = list()
tmp = 0

for file_name in files_data:
    tmp += files_data[file_name]['Цена']
portfolio_price = list(tmp)
```

```python
portfolio_profit = [(portfolio_price[i] - portfolio_price[i-1]) * 100 / portfolio_price[i-1]\
                    for i in range(len(portfolio_price)-1)]
```

```python
portfolio_profit = np.sort(portfolio_profit)
portfolio_coef = int(len(portfolio_profit)*alpha)
VaR_st = -portfolio_profit[portfolio_coef]
print('VaR портфеля = {var} %'.format(var=VaR_st))
```

```python
# Доходность портфеля
```

```python
cut_cases = portfolio_profit[:portfolio_coef]
remain_cases = portfolio_profit[portfolio_coef:len(portfolio_profit)]
plt.figure(figsize=(15, 10))
plt.plot(range(len(cut_cases)), cut_cases, color='red')
plt.plot(range(len(cut_cases), len(remain_cases) + len(cut_cases)), remain_cases, color='green')
plt.xlabel('Дни, шт')
plt.ylabel('Доходность, %')
plt.grid()
plt.show()
```

```python
# Суммарный VaR
```

```python
import random

# T - период
# k - расчитываем изменение цены акции за k дней
T = len(portfolio_price)
print(T)
k = 2

mid_bid_ask_prices = dict.fromkeys(files_data.keys())
for company, data in files_data.items():
  mid_bid_ask_prices[company] = [random.uniform(price - 10, price + 10) for price in files_data[company]['Цена']]

print(mid_bid_ask_prices)
```

```python
U = dict.fromkeys(files_data.keys())
U_star = dict.fromkeys(files_data.keys())

T_ = 0
for company in files_data.keys():
    T_ = len(mid_bid_ask_prices[company])
    U[company] = np.log(mid_bid_ask_prices[company][k:])- np.log(mid_bid_ask_prices[company][:T_-k])
    U[company] = np.sort(U[company])
    U_star[company] = -U[company][int(alpha*(T_-k))]
    print('VaRsum_{name} = {val} %'.format(name=company, val=U_star[company]*100))
```

```python
# Суммарный VaR всего портфеля
```

```python
T_min = len(mid_bid_ask_prices['GAZP.csv'])
for file_name in files_data.keys():
    if (T_min > len(mid_bid_ask_prices[file_name])):
        T_min = len(mid_bid_ask_prices[file_name])
```

```python
mid_bid_ask_portfolio = 0

for file_name in files_data:
    mid_bid_ask_portfolio += np.array(mid_bid_ask_prices[file_name][:T_min])

U_portfolio = np.log(mid_bid_ask_portfolio[k:]) - np.log(mid_bid_ask_portfolio[:T_min-k])
U_portfolio = np.sort(U_portfolio)
U_portfolio_star = U_portfolio[int(alpha*(T_min-k))]
VaR_sum = -U_portfolio_star
```

```python
print('VaRsum портфеля = {val} %'.format(val=-U_portfolio_star*100))
```

```python
# Оценка ликвидности портфеля
```

```python
for company in files_data.keys():
    col = (U_star[company] + sorted_data[company][coef])*100
    ml = U_star[company] / (-sorted_data[company][coef])
    print ('COL_{company} = {val}%'.format(company=company, val=col))
    print ('ML_{company} = {val} \n'.format(company=company, val=ml))
```

```python
COL = VaR_sum - VaR_st
ML = VaR_sum / VaR_st
print('COL портфеля = {val}%'.format(val=COL))
print('ML портфеля = {val}'.format(val=ML))
```

```python
# Монте-Карло
```

```python

for file_name in files_data:
    prices = files_data[file_name]['Цена']
    returns = np.log(prices / prices.shift(1)).dropna()


    # # Шаг 2: Определение параметров
    confidence_level = 0.95
    num_simulations = 10000

    mean_return = returns.mean()
    std_dev = returns.std()

    simulated_returns = np.random.normal(mean_return, std_dev, (num_simulations, len(returns)))

    initial_investment = 1
    portfolio_values = initial_investment * (1 + simulated_returns).cumprod(axis=1)

    portfolio_losses = initial_investment - portfolio_values[:, -1]
    VaR = np.percentile(portfolio_losses, (1 - confidence_level) * 100)

    print(f'VaR at {confidence_level*100}% confidence level: {VaR}')
    plt.hist(portfolio_losses, bins=50, alpha=0.75)
    plt.axvline(VaR, color='r', linestyle='dashed', linewidth=2)
    plt.title('Distribution of Portfolio Losses')
    plt.xlabel('Loss')
    plt.ylabel('Frequency')
    plt.show()
```
